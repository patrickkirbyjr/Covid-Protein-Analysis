```python
# import the COVID-19 genome and print the sequence
# source: https://www.ncbi.nlm.nih.gov/nuccore/NC_045512.2
# tutorial: https://www.youtube.com/watch?v=r5OQ-4Zmo50

from Bio import SeqIO

genome = SeqIO.read('sequence.fasta', 'fasta')
DNA_sequence = genome.seq

print(DNA_sequence)
```

```python
# print the length of nucleotides (bases) in the genome (research shows this should be between 27-31 kb)

DNA_sequence_length = len(DNA_sequence)
DNA_sequence_length
```

```python
# view the distribution of nucleotides within the genome

import matplotlib.pyplot as plt

count_nucleotides = {
    "A": DNA_sequence.count("A"),
    "C": DNA_sequence.count("C"),
    "G": DNA_sequence.count("G"),
    "T": DNA_sequence.count("T"),
}

fig, ax = plt.subplots()
ax.bar(count_nucleotides.keys(), 
       count_nucleotides.values(), 
       color = ["cornflowerblue", "crimson", "gold", "mediumseagreen"])
ax.set_xlabel("Nucleotides")
ax.set_ylabel("Frequency")
ax.set_title("Nucleotide Fequency in COVID-19 Genome");
ax.tick_params(axis='x', labelsize=16);
```

```python
# determine the GC content of the genome

from Bio.SeqUtils import gc_fraction

gc_content = 100 * gc_fraction(DNA_sequence)
print(f"GC Content of COVID-19 Genome: {100 * gc_fraction(DNA_sequence):.2f}%")
```

```python
# convert DNA to RNA (transcription)

mRNA_sequence = DNA_sequence.transcribe()
print(mRNA_sequence)
```

```python
# compare first 50 bases of DNA and RNA

print(DNA_sequence[:50])
print(mRNA_sequence[:50])
```

```python
# translate from mRNA to amino acids

amino_acids = mRNA_sequence.translate()
print(len(amino_acids))
print(amino_acids)
```

```python
# determine the most common amino acids

from collections import Counter

Counter(amino_acids).most_common(10)
```

```python
# visualize the top amino acids

top_aa = Counter(amino_acids).most_common(11)
aa_counts = [count for letter, count in top_aa if letter != "*"]
aa_letters = [letter for letter, count in top_aa if letter != "*"]

fig, ax = plt.subplots()
ax.bar(aa_letters, aa_counts, color = ["cornflowerblue", "crimson", "gold", "mediumseagreen"])
ax.set_xlabel("Amino Acids")
ax.set_ylabel("Frequency")
ax.set_title("Amino Acid Frequency in COVID-19 Genome");
ax.tick_params(axis="x", labelsize=16);
```

```python
# protein analysis: split on stop codon

proteins = [str(i) for i in amino_acids.split("*")]
proteins = amino_acids.split("*")
print(proteins)
```

```python
# protein analysis: number of proteins found

protein_count = len(proteins)
protein_count
```

```python
# protein analysis: length of first amino acid sequences

import pandas as pd

df = pd.DataFrame({"Amino Acids": proteins })
df["Count"] = df["Amino Acids"].str.len()
df.head(10)
```

```python
# protein analysis: length of longest amino acid sequences

df.nlargest(n=10, columns="Count")
```

```python
# protein analysis: determine the number of functional proteins (greater than 20 amino acids)

functional_proteins = 0

for protein in proteins:
    if len(protein) < 20:
        continue
    functional_proteins += 1

functional_proteins
```

```python
# summary

print("Subject: COVID-19")
print("Length of DNA Sequence:", DNA_sequence_length)
print("GC Content:", round(gc_content, 2), "%")
print("Number of Proteins:", protein_count)
print("Number of Functional Proteins:", functional_proteins)
```

```python
# save largest protein for further analysis

top_5_proteins = sorted(proteins, key=len, reverse=True)

with open("largest_protein.fasta", "w") as file:
    file.write(f">protein\n{top_5_proteins[0]}")
```

```python
# use BLAST to search for protein

from Bio.Blast import NCBIWWW
from Bio import SearchIO, SeqIO

protein_sequence = SeqIO.read("largest_protein.fasta", "fasta")

result_handle = NCBIWWW.qblast("blastp", "pdb", protein_sequence.seq)

blast_records = SearchIO.read(result_handle, "blast-xml")
```

```python
# show top three matches from BLAST search

for blast_record in blast_records[:3]:
    print(f"Sequence ID: {blast_record.id}")
    print(f"Description: {blast_record.description}")
    print(f"E Value: {blast_record[0].evalue}")
    print(f"Bit Score: {blast_record[0].bitscore}")
    print(f"Alignment: {blast_record[0].aln}")
    print()
```
